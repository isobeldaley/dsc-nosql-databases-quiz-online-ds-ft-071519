## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

_______________________________________________________________________________________________________________________________

A SQL database is organised into tables, each of which can be linked to others through the use of primary/foreign keys.

A NoSQL database is not organised into tables.  An example would be MongoDB, which stores data in documents (which can be related to one another).

## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


Fault tolerance is the ability of a dataset to withstand a problem with one of the servers on which it is held.  Resilient Distributed Datasets are spread across multiple servers.  Each server contains one part of the dataset, but also backup(s) of other part(s).  This means, if one of the servers goes down, it is still possible to access the entire dataset.  For this reason, it is resilient, and has a high fault tolerance. 





## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Hadoop supports RDD.  When a query is run on Hadoop, it is effectively run simultaneously on mutiple servers (containing different parts of the dataset).  The results are then combined ('reduced') to give the total result.
